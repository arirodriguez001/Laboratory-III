```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Parmetros del sistema de Lorenz
sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0

# Funcin para las ecuaciones del sistema de Lorenz
def lorenz(X, t=0):
    x, y, z = X
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return np.array([dxdt, dydt, dzdt])

# Condiciones iniciales
X0 = [1.0, 0.0, 0.0]  # Valores iniciales de x, y, z
t = np.linspace(0, 40, 10000)  # Tiempo para la simulacin

# Integracin numrica (Mtodo de Runge-Kutta)
from scipy.integrate import odeint
sol = odeint(lorenz, X0, t)

# Graficando el Atractor de Lorenz
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(sol[:, 0], sol[:, 1], sol[:, 2], lw=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Atractor de Lorenz')

plt.show()
```

```python
# Parmetros del sistema de Hnon
a = 1.4
b = 0.3

# Condiciones iniciales
x0, y0 = 0, 0

# Nmero de iteraciones
iterations = 10000

# Arrays para almacenar los puntos
x_vals = np.zeros(iterations)
y_vals = np.zeros(iterations)

# Inicializacin de los valores
x_vals[0], y_vals[0] = x0, y0

# Generar la trayectoria del sistema de Hnon
for i in range(1, iterations):
    x_vals[i] = 1 - a * x_vals[i-1]**2 + y_vals[i-1]
    y_vals[i] = b * x_vals[i-1]

# Graficar el atractor de Hnon
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, '.', color='black', markersize=0.5)
plt.title('Atractor de Hnon')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Parmetros del sistema de Rssler
a = 0.2
b = 0.2
c = 5.7

# Condiciones iniciales
x0, y0, z0 = 0, 1, 1

# Tiempo y pasos de integracin
dt = 0.01
t_max = 100
t = np.arange(0, t_max, dt)

# Inicializacin de las variables
x = np.zeros(len(t))
y = np.zeros(len(t))
z = np.zeros(len(t))

# Condiciones iniciales
x[0], y[0], z[0] = x0, y0, z0

# Integracin de las ecuaciones de Rssler usando el mtodo de Euler
for i in range(1, len(t)):
    dx = -y[i-1] - z[i-1]
    dy = x[i-1] + a * y[i-1]
    dz = b + z[i-1] * (x[i-1] - c)

    # Actualizacin de las variables
    x[i] = x[i-1] + dx * dt
    y[i] = y[i-1] + dy * dt
    z[i] = z[i-1] + dz * dt

# Graficar el atractor de Rssler en 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, color='black', lw=0.5)
ax.set_title('Atractor de Rssler')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
