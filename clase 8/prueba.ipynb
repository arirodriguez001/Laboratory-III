```python
import numpy as np
import nysg_tools as ny
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,np.pi,1000)
y = np.sin(x)

def curvaaj(x,a,b,c):
    return a*x**2+b*x+c


p0 = [1,1,1]

p = ny.fit_lsq(x,y,curvaaj,p0=[1,1,1])
```

```python
p["params"][1]
```

```python
plt.plot(x,y)

plt.plot(x,curvaaj(x,*p["params"]))
```
